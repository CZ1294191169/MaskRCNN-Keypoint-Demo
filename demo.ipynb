```
import cv2 as cv 
import numpy as np
import matplotlib.pyplot as plt

def load(name):
    """
    Given an url of an image, downloads the image and
    returns a PIL image
    """
    image = cv.imread(name)
    # convert to BGR format
    image = np.array(image)
    return image
  
def imshow(img):
    plt.imshow(img[:, :, [2, 1, 0]])
    plt.axis("off")
```

```
image = load("media/ski.jpg")
imshow(image)
```
